```python
Q1. What are the characteristics of the tuples? Is tuple immutable?
In Python, a tuple is a collection of ordered, immutable elements. Some of the key characteristics of tuples include:
Tuples are ordered: The elements in a tuple are ordered and each element can be accessed by its index.
Tuples can contain mixed data types: Unlike lists, tuples can contain elements of different data types.
Tuples are immutable: Once a tuple is created, it cannot be changed. This means that you cannot add, remove, or modify elements in a tuple.
Tuples are faster than lists: Because tuples are immutable, they can be processed faster than lists.
Tuples can be used as keys in dictionaries: Because tuples are immutable, they can be used as keys in dictionaries. Lists cannot be used as keys in dictionaries because they are mutable.
So, yes, tuples are immutable in Python. Once a tuple is created, it cannot be modified. If you need to modify a tuple, you will need to create a new tuple with the desired modifications.

Q2. What are the two tuple methods in python? Give an example of each method. Give a reason why tuples have only two in-built methods as compared to Lists.
In Python, there are two built-in methods for tuples:
count(): This method returns the number of times a specified element appears in the tuple.
Example:
my_tuple = (1, 2, 3, 1, 4, 5, 1)
count_ones = my_tuple.count(1)
print(count_ones) # Output: 3
index(): This method returns the index of the first occurrence of a specified element in the tuple. If the element is not found in the tuple, a ValueError is raised.
Example:
my_tuple = (1, 2, 3, 1, 4, 5, 1)
index_four = my_tuple.index(4)
print(index_four) # Output: 4
Tuples have only two built-in methods as compared to lists because tuples are immutable and do not need as many methods as lists.
Lists have more methods because they are mutable and require more methods to modify their contents. 
Additionally, tuples are typically used for simple, fixed collections of data, whereas lists are used for more complex data structures that require more operations, 
such as adding or removing elements. Therefore, the limited number of methods for tuples is not a limitation, but rather a reflection of their intended use.

Q3. Which collection datatypes in python do not allow duplicate items? Write a code using a set to remove
duplicates from the given list.
List = [1, 1, 1, 2, 1, 3, 1, 4, 2, 1, 2, 2, 2, 3, 2, 4, 3, 1, 3, 2, 3, 3, 3, 4, 4, 1, 4, 2, 4, 3, 4, 4]
The set data type in Python does not allow duplicate items. When a set is created from a list, all duplicates are automatically removed. Here's an example code using a set to remove duplicates from the given list:
my_list = [1, 1, 1, 2, 1, 3, 1, 4, 2, 1, 2, 2, 2, 3, 2, 4, 3, 1, 3, 2, 3, 3, 3, 4, 4, 1, 4, 2, 4, 3, 4, 4]
unique_items = set(my_list)
print(unique_items)
Output:
{1, 2, 3, 4}
In the code above, we first define the list my_list that contains duplicates. Then, we create a set unique_items from the list using the set() function. The resulting set contains only unique elements from the original list, with duplicates automatically removed.
Note that the order of the elements in the resulting set is not guaranteed to be the same as the order of the original list, since sets are unordered collections.

Q4. Explain the difference between the union() and update() methods for a set. Give an example of each method.
In Python, both the union() and update() methods can be used to combine two or more sets. However, there is a difference between these two methods:

union() method: This method returns a new set that contains all unique elements from all sets that were included in the union. The original sets are not modified. The union() method can take one or more sets as arguments.
Example:
set1 = {1, 2, 3}
set2 = {2, 3, 4}
set3 = {3, 4, 5}
union_set = set1.union(set2, set3)
print(union_set) # Output: {1, 2, 3, 4, 5}
In the example above, we first define three sets set1, set2, and set3. Then, we use the union() method to combine these sets into a new set union_set that contains all unique elements from all three sets.

update() method: This method updates the original set with all unique elements from one or more sets. The update() method can take one or more sets as arguments. The original set is modified.
Example:

set1 = {1, 2, 3}
set2 = {2, 3, 4}
set3 = {3, 4, 5}
set1.update(set2, set3)
print(set1) # Output: {1, 2, 3, 4, 5}
In the example above, we first define three sets set1, set2, and set3. Then, we use the update() method to add all unique elements from set2 and set3 to set1. The resulting set is stored back in set1.
So, the main difference between the union() and update() methods is that union() returns a new set, while update() modifies the original set. Additionally, union() can take multiple sets as arguments, while update() can only take one or more sets as arguments.

Q5. What is a dictionary? Give an example. Also, state whether a dictionary is ordered or unordered.
In Python, a dictionary is a collection data type that is used to store data in the form of key-value pairs. Each key-value pair in a dictionary is separated by a colon (:), and the pairs are separated by commas (,). Dictionaries are enclosed in curly braces ({}).

Here's an example of a dictionary that stores information about a person's name, age, and city:
person = {'name': 'John', 'age': 30, 'city': 'New York'}
In the example above, the dictionary person has three key-value pairs: 'name': 'John', 'age': 30, and 'city': 'New York'. The keys are strings ('name', 'age', and 'city'), and the values can be of any data type ('John' is a string, 30 is an integer, and 'New York' is a string).
One important thing to note about dictionaries is that they are unordered collections, which means that the key-value pairs do not have any specific order. The order of the pairs in a dictionary is not guaranteed to be the same as the order in which they were defined. However, starting from Python 3.7, dictionaries retain the order of insertion, so in practice, dictionaries behave like ordered collections in modern versions of Python.

Q6. Can we create a nested dictionary? If so, please give an example by creating a simple one-level nested dictionary.
Yes, we can create a nested dictionary in Python. A nested dictionary is a dictionary inside another dictionary, where the values of the outer dictionary are themselves dictionaries.

Here's an example of a simple one-level nested dictionary:

students = {'Alice': {'age': 20, 'major': 'Computer Science'},
            'Bob': {'age': 21, 'major': 'Mathematics'},
            'Charlie': {'age': 19, 'major': 'Physics'}}
In the example above, the students dictionary has three keys: 'Alice', 'Bob', and 'Charlie'. The values of each key are themselves dictionaries with two key-value pairs: 'age' and 'major'.

For example, if we want to access Alice's age, we can do so using the following code:

alice_age = students['Alice']['age']
print(alice_age) # Output: 20
In the code above, we first access the 'Alice' key in the students dictionary, which returns a nested dictionary with the 'age' and 'major' keys. Then, we access the 'age' key of the nested dictionary to retrieve Alice's age.

Q7. Using setdefault() method, create key named topics in the given dictionary and also add the value of
the key as this list ['Python', 'Machine Learningâ€™, 'Deep Learning']
dict1 = {'language' : 'Python', 'course': 'Data Science Masters'}
Here's an example code that uses the setdefault() method to create a key named topics in the given dictionary and assigns it the value ['Python', 'Machine Learning', 'Deep Learning']:

dict1 = {'language': 'Python', 'course': 'Data Science Masters'}
dict1.setdefault('topics', ['Python', 'Machine Learning', 'Deep Learning'])
print(dict1)
Output:

{'language': 'Python', 'course': 'Data Science Masters', 'topics': ['Python', 'Machine Learning', 'Deep Learning']}
In the code above, we use the setdefault() method to create a new key named 'topics' in the dict1 dictionary if it does not already exist. If the key already exists, the method does not modify the dictionary. We set the value of the 'topics' key to ['Python', 'Machine Learning', 'Deep Learning']. Finally, we print the modified dict1 dictionary to verify that the new key has been added.

Q8. What are the three view objects in dictionaries? Use the three in-built methods in python to display these three view objects for the given dictionary.
In Python, dictionaries have three view objects that provide dynamic views of the dictionary's keys, values, and key-value pairs:

dict.keys(): returns a view object that contains the keys of the dictionary.
dict.values(): returns a view object that contains the values of the dictionary.
dict.items(): returns a view object that contains the key-value pairs of the dictionary.
Here's an example code that demonstrates how to use these three view objects to display the keys, values, and items of a dictionary:

my_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4}

# Display the keys of the dictionary
print("Keys:", my_dict.keys())

# Display the values of the dictionary
print("Values:", my_dict.values())

# Display the items of the dictionary
print("Items:", my_dict.items())
Output:

Keys: dict_keys(['a', 'b', 'c', 'd'])
Values: dict_values([1, 2, 3, 4])
Items: dict_items([('a', 1), ('b', 2), ('c', 3), ('d', 4)])
In the code above, we define a dictionary named my_dict that has four key-value pairs. Then, we use the keys(), values(), and items() methods to display the view objects that correspond to the keys, values, and items of the dictionary. Note that the view objects are displayed as dict_keys, dict_values, and dict_items, respectively.
```
